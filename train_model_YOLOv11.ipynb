```python
import os
from ultralytics import YOLO
from roboflow import Roboflow
from google.colab import drive, userdata

# 1. Mount Drive explicitly
drive.mount('/content/drive')

# 2. SAFETY CHECK: Verify Drive is actually accessible before starting
if not os.path.exists('/content/drive/MyDrive'):
    raise RuntimeError("CRITICAL ERROR: Google Drive is not mounted! processing will be lost.")



apiKey = userdata.get('ROBOFLOW_API_KEY')
rf = Roboflow(api_key=apiKey)
project = rf.workspace("projects-vnzqz").project("tennis-ball-obj-det-cyjrp-thstm-vl2fz")
version = project.version(1)
dataset = version.download("yolov11")


model = YOLO("/content/drive/MyDrive/Coding Stuff/Tennis Serve Tracker/best.pt") 

# 3. Train with safety settings
results = model.train(
    data="/content/Tennis-Ball-Obj-Det-1/data.yaml", 
    epochs=100, 
    imgsz=640, 
    
    # Path settings
    project="/content/drive/MyDrive/Coding Stuff/Tennis Serve Tracker/YOLOv11_Training_Results", 
    name="run1",
    resume=True,
    # Safety settings
    save=True,      # Saves last.pt and best.pt
    save_period=5   # Saves a permanent backup file every 5 epochs
)
```
